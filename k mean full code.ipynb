```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\User\Downloads\Cancer_Data.csv")
df
```

```python
df = df.drop(columns= ["diagnosis" , "Unnamed: 32"])
df
```

```python
def initialize_centroids(n) :
    indices = np.random.choice(df.shape[0] , n , replace=False)
    return df.iloc[indices].to_numpy()

# c = initialize_centroids(2)
```

```python
def compute_distances(x_df, centroids):
    x_df = pd.DataFrame(x_df)
    x = x_df.to_numpy()
    d = [] 
    for i in range(len(x)) :
        temp =[]        
        for j in range(len(centroids)) :
            diff = np.subtract(x[i] , centroids[j]) 
            r = np.sqrt(np.sum(diff**2))
            temp.append(r)
        d.append(temp.index(min(temp))) 
    d = np.array(d)
    return d

# d = compute_distances(df , c )
```

```python
def update_centroid(x_df , centroid ,d) :
    x_df = pd.DataFrame(x_df)
    x = x_df.to_numpy()
    new_centroied = []

    for i in range(len(centroid)):
        cluster_points = x[d == i] 
        if len(cluster_points) > 0:
            mean_point = np.mean(cluster_points, axis=0)
            new_centroied.append(mean_point)    
    return np.array(new_centroied)

# u = update_centroid( df, c)
# u
```

```python
itr = 20
c = initialize_centroids(2)
d = compute_distances(df , c )
u = update_centroid( df, c , d)
print(c)
for i in range(itr) :
    d = compute_distances( df , u )
    u = update_centroid( df, u , d)

    print("itr num  : " , i , u)



```
